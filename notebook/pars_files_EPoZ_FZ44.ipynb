```python
from transformers import AutoTokenizer, T5ForConditionalGeneration

model_name = "IlyaGusev/rut5_base_sum_gazeta"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

def summary_data(article_text):
  input_ids = tokenizer(
      [article_text],
      max_length=600,
      add_special_tokens=True,
      padding="max_length",
      truncation=True,
      return_tensors="pt"
  )["input_ids"]

  output_ids = model.generate(
      input_ids=input_ids,
      no_repeat_ngram_size=4
  )[0]

  return tokenizer.decode(output_ids, skip_special_tokens=True)

```

# 44 ФЗ

```python
from collections import defaultdict
from docx import Document


def_dict = defaultdict(list)
file_path = "/content/44 ФЗ.docx"

def read_docx_lines(file_path):
    doc = Document(file_path)
    lines = []
    for paragraph in doc.paragraphs:
        lines.extend(paragraph.text.split('\n'))
    return [line.strip() for line in lines if line.strip()]

title = ""
count = 0
lines = read_docx_lines(file_path)
for line in lines:
    if "Статья" in line[:15]:
        title = line
        print(f"{count} статья обработана")
        count+=1
    else:
        def_dict[title].append(line)
```

```python
import csv


csv_file_path = "output.csv"
with open(csv_file_path, mode='w', newline='', encoding='utf-8') as csv_file:
    # Создаем объект writer
    csv_writer = csv.writer(csv_file)

    # Записываем заголовки
    csv_writer.writerow(['query', 'answer'])

    # Записываем данные
    for key, value in def_dict.items():
        string = ' '.join(value)
        csv_writer.writerow([key, string])

print(f'Data saved to {csv_file_path}')
```

```python
from collections import defaultdict
from docx import Document

def_dict = defaultdict(list)
file_path = "/content/ЕПоЗ_май 2023.docx"

# Ваш список заголовков
headings_list = [
    "Правовая основа закупочной деятельности Корпорации и организаций Корпорации",
    "Цели и принципы закупочной деятельности",
    "Информационное обеспечение",
    "Субъекты закупочной деятельности",
    "Органы управления закупочной деятельностью, их функции и полномочия",
    "Способы закупок, формы закупок, условия их применения",
    "Дополнительные элементы закупок",
    "Планирование закупок",
    "Процесс подготовки к проведению закупки",
    "Порядок проведения процедур закупки",
    "Порядок проведения конкурентной процедуры закупки",
    "Порядок проведения неконкурентной процедуры закупки ",
    "Требования к ЭТП/ЗЭТП",
    "Особенности проведения конкурентных способов закупки в бумажной форме",
    "Особенности принятия решений, установления требований и/или порядка проведения процедур закупок в отношении отдельных видов закупаемой продукции, рынков, закупочных ситуаций",
    "Заключение договоров",
    "Исполнение договора",
    "Обжалование действий (бездействия) заказчика, организатора закупки, закупочной комиссии"


]

def read_docx_lines(file_path, headings_list):
    doc = Document(file_path)
    lines = []
    current_heading = None

    for paragraph in doc.paragraphs:
        text = paragraph.text.strip()

        # Проверяем, является ли абзац заголовком из вашего списка
        if text in headings_list:
            current_heading = text
            continue

        # Добавляем текст в словарь по текущему заголовку из списка
        if current_heading:
            def_dict[current_heading].append(text)

    return def_dict

# Вызываем функцию и выводим результат
result_dict = read_docx_lines(file_path, headings_list)
for heading, lines in result_dict.items():
    print(f"{heading}: {lines}")

len(result_dict.keys())
```

```python
import csv


csv_file_path = "EPoZ.csv"
with open(csv_file_path, mode='w', newline='', encoding='utf-8') as csv_file:
    # Создаем объект writer
    csv_writer = csv.writer(csv_file)

    # Записываем заголовки
    csv_writer.writerow(['query', 'answer'])

    # Записываем данные
    for key, value in result_dict.items():
        string = ' '.join(value)
        csv_writer.writerow([key, string])

print(f'Data saved to {csv_file_path}')
```
